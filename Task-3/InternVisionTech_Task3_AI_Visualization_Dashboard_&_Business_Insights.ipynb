##KPI Summary

The KPI (Key Performance Indicator) Dashboard provides a quick overview of the overall sales performance. It highlights important business metrics such as total orders, total sales, total profit, and average discount to support data-driven decision-making.

```python
# ============================================
# KPI SUMMARY CARDS (Professional Dashboard)
# ============================================

import matplotlib.pyplot as plt

# -----------------------------
# KPI Calculations
# -----------------------------

total_orders = len(df)
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
avg_discount = df["Discount_Percentage"].mean()

# -----------------------------
# KPI Information
# -----------------------------

titles = [
    "Total Orders",
    "Total Sales",
    "Total Profit",
    "Avg Discount"
]

values = [
    f"{total_orders}",
    f"₹{total_sales:,.0f}",
    f"₹{total_profit:,.0f}",
    f"{avg_discount:.1f}%"
]

# Card Background Colors
colors = [
    "#E8F4FD",   # Light Blue
    "#EAF8EC",   # Light Green
    "#FFF4E5",   # Light Orange
    "#F8E8F5"    # Light Pink
]

# -----------------------------
# Create Dashboard
# -----------------------------

fig, axes = plt.subplots(1, 4, figsize=(18, 4))

fig.patch.set_facecolor("white")

for ax, title, value, color in zip(axes, titles, values, colors):

    ax.set_facecolor(color)

    ax.set_xticks([])
    ax.set_yticks([])

    for spine in ax.spines.values():
        spine.set_visible(False)

    # KPI Title
    ax.text(
        0.5,
        0.72,
        title,
        ha="center",
        va="center",
        fontsize=14,
        fontweight="bold",
        color="#2E4053"
    )

    # KPI Value
    ax.text(
        0.5,
        0.38,
        value,
        ha="center",
        va="center",
        fontsize=24,
        fontweight="bold",
        color="#1B2631"
    )

    # Bottom Line
    ax.axhline(
        y=0.08,
        xmin=0.15,
        xmax=0.85,
        linewidth=4,
        color="#5D6D7E"
    )

# Dashboard Title
plt.suptitle(
    "Sales KPI Dashboard",
    fontsize=20,
    fontweight="bold"
)

plt.tight_layout(rect=[0, 0, 1, 0.92])

plt.show()
```

##**Data Visualization** ##

##1. Bar Chart

This bar chart compares the total sales generated by each product category, helping identify the best-performing categories.

```python
plt.figure(figsize=(8,5))

sales_by_category = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

sales_by_category.plot(kind="bar")

plt.title("Total Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=30)

plt.show()
```

##2. Pie Chart

This pie chart illustrates the percentage contribution of each region to the overall sales.

```python
plt.figure(figsize=(7,7))

region_sales = df.groupby("Region")["Sales"].sum()

plt.pie(
    region_sales,
    labels=region_sales.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Sales Distribution by Region")

plt.show()
```

##3. Histogram

The histogram displays the distribution of sales values to understand how sales are spread across orders.

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=10)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

##4. Box Plot

The box plot is used to identify outliers in the Profit column and understand the spread of profit values.

```python
plt.figure(figsize=(8,4))

sns.boxplot(x=df["Profit"])

plt.title("Profit Distribution")

plt.show()
```

##5. Scatter Plot

This scatter plot shows the relationship between Sales and Profit to identify possible trends.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Sales",
    y="Profit",
    hue="Category"
)

plt.title("Sales vs Profit")

plt.show()
```

##6. Correlation Heatmap

The heatmap visualizes the correlation between numerical variables, helping identify strong positive or negative relationships.

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

##Business Dashboard

The Business Dashboard combines multiple visualizations into a single view to provide a comprehensive overview of sales performance. It helps identify trends, compare categories, analyze customer behavior, and support data-driven business decision-making.

```python
# ============================================
# BUSINESS DASHBOARD
# ============================================

import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(3, 2, figsize=(18, 15))

fig.suptitle(
    "Sales Performance Dashboard",
    fontsize=20,
    fontweight="bold"
)

# ============================================
# 1. BAR CHART
# ============================================

sales_by_category = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

sales_by_category.plot(
    kind="bar",
    ax=axes[0,0]
)

axes[0,0].set_title("Sales by Category")
axes[0,0].set_xlabel("Category")
axes[0,0].set_ylabel("Sales")
axes[0,0].tick_params(axis='x', rotation=25)

# ============================================
# 2. PIE CHART
# ============================================

region_sales = df.groupby("Region")["Sales"].sum()

axes[0,1].pie(
    region_sales,
    labels=region_sales.index,
    autopct="%1.1f%%",
    startangle=90
)

axes[0,1].set_title("Sales Distribution by Region")

# ============================================
# 3. HISTOGRAM
# ============================================

axes[1,0].hist(
    df["Sales"],
    bins=10
)

axes[1,0].set_title("Sales Distribution")
axes[1,0].set_xlabel("Sales")
axes[1,0].set_ylabel("Frequency")

# ============================================
# 4. BOX PLOT
# ============================================

sns.boxplot(
    x=df["Profit"],
    ax=axes[1,1]
)

axes[1,1].set_title("Profit Distribution")

# ============================================
# 5. SCATTER PLOT
# ============================================

sns.scatterplot(
    data=df,
    x="Sales",
    y="Profit",
    hue="Category",
    ax=axes[2,0]
)

axes[2,0].set_title("Sales vs Profit")

# ============================================
# 6. CORRELATION HEATMAP
# ============================================

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm",
    ax=axes[2,1]
)

axes[2,1].set_title("Correlation Heatmap")

plt.tight_layout()

plt.show()
```

##Trend Analysis

Trend analysis helps understand how sales change over time. By analyzing monthly sales performance, businesses can identify seasonal patterns, monitor growth, and make better strategic decisions.

```python
# ============================================
# TREND ANALYSIS
# ============================================

# Convert Order_Date to datetime format
df["Order_Date"] = pd.to_datetime(
    df["Order_Date"],
    format="%d-%m-%Y"
)

# Extract Month Name
df["Month"] = df["Order_Date"].dt.strftime("%b")

# Monthly Sales
monthly_sales = (
    df.groupby("Month")["Sales"]
    .sum()
    .reindex(["Jan", "Feb", "Mar", "Apr", "May", "Jun"])
)

plt.figure(figsize=(10,5))

plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker='o',
    linewidth=3
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.grid(alpha=0.3)

plt.show()

print("\n")
print("Insight:The line chart illustrates monthly sales performance, helping identify business growth trends and seasonal variations.")
```

## Business Insights

The following insights are generated automatically from the sales dataset based on key business metrics and visualizations. These insights help identify business opportunities and support data-driven decision-making.

```python
# ============================================
# BUSINESS INSIGHTS REPORT
# ============================================

print("="*75)
print("📊 BUSINESS INSIGHTS REPORT")
print("="*75)

# ----------------------------
# KPI Summary
# ----------------------------

total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = len(df)
total_quantity = df["Quantity"].sum()

print(f"🛒 Total Orders           : {total_orders}")
print(f"💰 Total Sales            : ₹{total_sales:,.2f}")
print(f"📈 Total Profit           : ₹{total_profit:,.2f}")
print(f"📦 Total Quantity Sold    : {total_quantity}")
print(f"🏷 Average Discount       : {df['Discount_Percentage'].mean():.2f}%")

print("\n" + "="*75)
print("📌 KEY BUSINESS INSIGHTS")
print("="*75)

# ----------------------------
# Highest Sales Category
# ----------------------------

category_sales = df.groupby("Category")["Sales"].sum()

best_category = category_sales.idxmax()
best_category_sales = category_sales.max()

print(f"✅ '{best_category}' generated the highest sales (₹{best_category_sales:,.2f}).")

# ----------------------------
# Highest Sales Region
# ----------------------------

region_sales = df.groupby("Region")["Sales"].sum()

best_region = region_sales.idxmax()

print(f"✅ The '{best_region}' region recorded the highest overall sales.")

# ----------------------------
# Best Customer Type
# ----------------------------

customer_sales = df.groupby("Customer_Type")["Sales"].sum()

best_customer = customer_sales.idxmax()

print(f"✅ '{best_customer}' customers contributed the highest revenue.")

# ----------------------------
# Best Selling Product
# ----------------------------

product_sales = df.groupby("Product")["Sales"].sum()

best_product = product_sales.idxmax()

print(f"✅ '{best_product}' was the highest-selling product.")

# ----------------------------
# Most Profitable Category
# ----------------------------

category_profit = df.groupby("Category")["Profit"].sum()

best_profit_category = category_profit.idxmax()

print(f"✅ '{best_profit_category}' generated the highest total profit.")

# ----------------------------
# Highest Average Discount
# ----------------------------

highest_discount = df["Discount_Percentage"].max()

print(f"✅ The maximum discount offered was {highest_discount}%.")

# ----------------------------
# Average Order Value
# ----------------------------

average_order_value = df["Sales"].mean()

print(f"✅ The average order value was ₹{average_order_value:,.2f}.")

# ----------------------------
# Profit Margin
# ----------------------------

profit_margin = (total_profit / total_sales) * 100

print(f"✅ Overall business profit margin was {profit_margin:.2f}%.")

print("="*75)
print("🎉 Business Insights Generated Successfully")
print("="*75)
```
