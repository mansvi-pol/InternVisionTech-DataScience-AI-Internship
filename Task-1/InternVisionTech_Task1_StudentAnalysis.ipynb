```python
# import : loads a library into python
# pandas : a library used to work with tables(datasets)
# numpy : used for mathematical calculations
# matplotlib : used to create graphs , pyplot contains graph functions
# seaborn : graph library used for beautiful and professional-looking graphs

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# it will open the file named student_data.csv
df = pd.read_csv("student_data.csv")
```

```python
# it will displays the entire dataset
df
```

```python
# shows no of rows,no of columns,column names,data type,memory usage
df.info()
```

```python
# Checks whether any cell is empty(checks missing values)
df.isnull().sum()
```

```python
#generate summary statistics(calculates mean,minimum,maximum,standard deviation,count)
df.describe()
```

```python
## Data Analysis ##

# calulates average marks

average_marks = df["Marks Obtained"].mean()
print("Average Marks:", average_marks)

# finds highest and lowest marks

highest = df["Marks Obtained"].max()
print("\nHighest Marks:", highest)
lowest = df["Marks Obtained"].min()
print("\nLowest Marks:",lowest)

# department-wise Performance

print("\nDepartment-wise Performance:\n")
department = df.groupby("Department")["Marks Obtained"].mean()
print(department)


# attendance Vs marks analysis

correlation = df["Attendance Percentage"].corr(df["Marks Obtained"])
print("\nAttendance Vs Marks analysis :", correlation)

# study hours Vs marks analysis

study = df["Study Hours"].corr(df["Marks Obtained"])
print("\nStudy hours Vs Marks analysis :", study)

# Top 10 performing students

print("\nTop 10 performing students:\n")
top10 = df.sort_values(by="Marks Obtained", ascending=False)
top10.head(10)
```

```python
## Data Visualization ##

# Bar chart

department.plot(kind="bar")                         # creates a bar chart
plt.title("Department Wise Average Marks")          # adds the graph title
plt.xlabel("Department")                            # names the x-axis
plt.ylabel("Average Marks")                         # names the y-axis
plt.show()                                          # displays the graph
```

```python
# Pie Chart

df["Department"].value_counts().plot(              #value_counts()counts students in each department
kind="pie",                                        #creates a pie chart
autopct="%1.1f%%")                                 #siplays percentage
plt.show()
```

```python
# Histogram : showing the distribution of marks

plt.hist(df["Marks Obtained"], bins=10)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Students")
plt.show()
```

```python
# Line Chart : showing how marks change across the students

plt.plot(df["Student ID"], df["Marks Obtained"])
plt.xlabel("Student ID")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.show()
```

```python
# Scatter Plot(gives basic apperance) : shows the relation between two numerical variables ("Study Hours vs Marks")

plt.scatter(df["Study Hours"], df["Marks Obtained"])             # creates a scatter plot by giving data for x & y axis by df
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")                                # graph title
plt.show()                                                       # displays the graph
```

```python
# Seaborn Scatter Plot(more professional apperance)

sns.scatterplot(                  # creates scatter plot using seaborn
x="Study Hours",
y="Marks Obtained",
data=df
)
plt.show()
```
